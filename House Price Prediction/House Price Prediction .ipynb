# House Price Prediction

## Linear Regression - using single variable

**In this project we are going to use Linear Regression Algorithm to predict the price of the house based on the amount of land that is provided.**

```python
# Importing the libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
dataset = pd.read_csv("house_price.csv")
dataset.head()
```

```python
# Extract the shape of the dataset to find the rows and columns
print(dataset.shape)
```

```python
# Dividing the data into independent and dependent variables
x = pd.DataFrame(dataset['area'])
y = pd.DataFrame(dataset['price'])
```

```python
# Data Visualization 
plt.scatter(x,y)
x_label = "Area"
y_label = "Price"
plt.show()
```

```python
sns.pairplot(dataset)
```

```python
# Split the data for training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Shape of the train and test sets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Train the algorithm
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predicting the test set results
x = 1500
LandArea = [[x]]
predicted_result = model.predict(LandArea)
print(predicted_result)
```

**Algorithm**

Steps to predict the price of the house mathematically using theory calculation:
- Plot x and y
- Finding Mean for X
- Finding Mean of Y
- Drawing vertical line from the point of mean on the x-axis
- Drawing horizontal line from the point of mean on the y-axis
- Mark the point of intersection of 2 lines
- Calculate x-x'
- Calculate y-y'
- Calculate (x-x')^2
- Now calculate the slope m using the formula ((x-x')(y-y'))/(x-x')^2
- Intercept c = y'-m*x' gives the value of c
- Substitute m and c in straight line equation



#### Let's cross check the model by theory calculation

y = m * x + c (m is the coefficient and c is the intercept)

```python
# Calculating the slope coefficient
m = model.coef_
print(m)
```

```python
# Calculating the intercept
c = model.intercept_
print(c)
```

```python
#Calculating the equation
y=m*x+c
print(y)
```

*Hence the price obtained after using Machine Learning Linear Regression is the same as the price obtained by the theory calculation done mathematically. This is an accurate working model to predict the price of any house provided the land available per square feet.*
