# Exam Mark Prediction

## Linear Regression - using Multiple variables

**In this project we are going to use Linear Regression Algorithm to predict the Marks scored by a student based on the Number of Courses and the Study Time of the student that is provided.**

```python
# Importing the libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the dataset
dataset = pd.read_csv("Student_Marks.csv")
dataset.head()
```

```python
# Extract the shape of the dataset to find the rows and columns
print(dataset.shape)
```

```python
# Finding null values from features x
dataset.columns[dataset.isna().any()]
```

```python
# Removing null values from features X
dataset.time_study = dataset.time_study.fillna(dataset.time_study.mean())
dataset
```

```python
# Segregate Dataset into input x and output y
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Data Visualization using Matplotlib
x1 = dataset.iloc[:,0].values
x2 = dataset.iloc[:,1].values
a = plt.subplot()
a.plot(x1,y,label="number_courses")
a.plot(x2,y,label="time_study")
plt.legend()
plt.show()
```

```python
# Data Visulaization using Seaborn
sns.pairplot(dataset)
```

```python
# Split the data for training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Shape of the train and test sets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Train the algorithm
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predicting the test set results
x_val = [[2,5.999]]
y_pred = model.predict(x_val)
print(y_pred)
```

```python
# Calculating the accuracy of the model
r2_score = model.score(x_train, y_train)
print(r2_score*100,'%')
```
